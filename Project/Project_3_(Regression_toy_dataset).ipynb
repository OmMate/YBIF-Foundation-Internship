# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Generate the dataset

```python
# without coeficient of underline model
from sklearn.datasets import make_regression
X,y= make_regression(n_samples=500,n_features=5,coef=False,bias=12,noise=10,random_state=2529)
```

```python
# with coeff of underline model
from sklearn.datasets import make_regression
X,y,w= make_regression(n_samples=500,n_features=5,coef=True,bias=12,noise=10,random_state=2529)
```

```python
X.shape, y.shape
```

```python
w
```

# Get the first 5 rows of y and X

```python
X[0:5]
```

```python
y[0:5]
```

# Splitting Train and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

# Get Intercept and Coefficients

```python
lr.intercept_
```

```python
lr.coef_
```

# Model Prediction

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
y_pred.shape
```

# Get Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
